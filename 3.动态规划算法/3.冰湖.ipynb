```python
import gym
from matplotlib import pyplot as plt
%matplotlib inline

#
#is_slippery
#map_name,8x8
#desc
env = gym.make('FrozenLake-v1',
               render_mode='rgb_array',
               is_slippery=False,
               map_name='4x4',
               desc=['SFFF', 'FHFH', 'FFFH', 'HFFG'])
env.reset()

#P
env = env.unwrapped


#
def show():
    plt.imshow(env.render())
    plt.show()


show()
```

```python
import numpy as np

#
values = np.zeros(16)

#
pi = np.ones([16, 4]) * 0.25

#
algorithm = ''
algorithm = ''

values, pi
```

```python
#qsa
def get_qsa(state, action):
    value = 0.0

    #
    for prop, next_state, reward, over in env.P[state][action]:

        #,values,
        next_value = values[next_state] * 0.9

        #0
        if over:
            next_value = 0

        #reward,
        next_value += reward

        #,
        next_value *= prop

        value += next_value

    return value


get_qsa(0, 0)
```

```python
#
def get_values():
    #values,
    new_values = np.zeros([16])

    #
    for state in range(16):

        #4
        action_value = np.zeros(4)

        #
        for action in range(4):
            action_value[action] = get_qsa(state, action)

        if algorithm == '':
            #
            action_value *= pi[state]
            #,
            new_values[state] = action_value.sum()

        if algorithm == '':
            #
            new_values[state] = action_value.max()

    return new_values


get_values()
```

```python
#
def get_pi():
    #,
    new_pi = np.zeros([16, 4])

    #
    for state in range(16):

        #4
        action_value = np.zeros(4)

        #
        for action in range(4):
            action_value[action] = get_qsa(state, action)

        #state
        count = (action_value == action_value.max()).sum()

        #
        for action in range(4):
            if action_value[action] == action_value.max():
                new_pi[state, action] = 1 / count
            else:
                new_pi[state, action] = 0

    return new_pi


get_pi()
```

```python
#
for _ in range(10):
    for _ in range(100):
        values = get_values()
    pi = get_pi()

values, pi
```

```python
#
def print_pi():
    #
    for row in range(4):

        line = ''

        for col in range(4):
            state = row * 4 + col

            if (row == 1 and col == 1) or (row == 1 and col == 3) or (
                    row == 2 and col == 3) or (row == 3 and col == 0):
                line += ''
                continue

            if row == 3 and col == 3:
                line += ''
                continue

            #line += ''
            line += ''[pi[state].argmax()]

        print(line)


print_pi()
```
